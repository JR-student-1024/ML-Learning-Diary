**Step_1** Create the variables x_train and y_train. The data is stored in one-dimensional NumPy arrays.

**Step_4** Plot the data  
Use the scatter() function from the Matplotlib library to plot these two points.

```python
#help(plt.scatter)
plt.scatter(x_train, y_train, marker = 'x', c = 'red')
#help(plt.title)
plt.title('Housing Price Related to Area')
plt.xlabel('Area (in 1000 sqrf)')
plt.ylabel('Price (in 1000s of dollars)');
```

**Step_5** Since the function expression for the linear regression model is$f_{(w,b)}(x^{(i)}) = wx^{(i)} + b$  
we can get different lines with different values of w and b.

```python
w = 200
b = 100
```

**Step_6** Use a function to generate predicted values.

```python
def compute_model_output(x, w, b):
    '''
    x(ndarray, (m,)), ndarray means x is an n-dimension array. (m,) means that x is 1D array now.
    w,b(scaller)
    '''
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

**Step_7** Plot the output.

```python
prediction = compute_model_output(x_train, w, b)
#help(plt.plot)

plt.plot(x_train, prediction, c = 'b', label = 'Prediction')
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Train Value')

plt.title('Housing Price Related to Area')
plt.xlabel('Area (in 1000 sqrf)')
plt.ylabel('Price (in 1000s of dollars)')
plt.legend()
plt.show()
```

By adjusting different values of w and b, the line is fitted to the data.
